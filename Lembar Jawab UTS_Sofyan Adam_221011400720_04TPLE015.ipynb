```python
def sort_company():
    # Extract company details from company_detail_list
    company_details = list(company_detail_list.values())

    # Sort company details by domain in reverse order
    sorted_company_details = sorted(company_details, key=lambda x: x['domain'], reverse=True)

    # Extract required fields (name and domain) from sorted company details
    result = [{'name': company['name'], 'domain': company['domain']} for company in sorted_company_details]

    return result

print(sort_company())
```

```python
# Extract company details and sort by domain in reverse order
company_details = [{'name': company_detail['name'], 'domain': company_detail['domain']} for company_name, company_detail in company_detail_list.items()]
sorted_companies = sorted(company_details, key=lambda x: x['domain'], reverse=True)

# Print the sorted companies
for company in sorted_companies:
    print(company)
```

```python
def get_company_domain():
    for company_name, company_detail in company_detail_list.items():
        domain = company_detail['domain']
        num_clients = len(company_detail['clients'])
        print(f"{company_name}: {domain}, relation: {num_clients} clients")

get_company_domain()
```

```python
def get_employees():
    employees = []

    for employee_name, employee_detail in employee_detail_list.items():
        full_name = employee_detail['full_name']
        company = employee_detail['company']

        company_domain = company_detail_list[company]['domain']

        employee_info = {
            'full_name': full_name,
            'company': company,
            'domain': company_domain
        }

        employees.append(employee_info)

    return employees

print(get_employees())
```

```python
def get_employees_by_company():
    companies_with_employees = []

    for company, company_detail in company_detail_list.items():
        employees = [employee['full_name'] for employee in employee_detail_list.values() if employee['company'] == company]

        company_info = {
            'company': company,
            'employees': employees
        }

        companies_with_employees.append(company_info)

    return companies_with_employees

print(get_employees_by_company())
```
